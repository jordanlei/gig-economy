# Simple Models

```python
from sklearn import tree
from sklearn.feature_extraction import DictVectorizer
```

```python
def split_data(df, train_split = 0.8):
    length = df.shape[0]
    myrange = np.arange(length)
    np.random.shuffle(myrange)
    train_split = int(train_split * length)
    return df.iloc[myrange[:train_split]].reset_index(drop = True), df.iloc[myrange[train_split:]].reset_index(drop = True)

def labels_data(df, label = "price"):
    x = df[list(set(df.columns) - set([label]))]
    y = df[label]
    return x, y

def accuracy(y, pred):
    return (y == pred)/len(pred)

def rmse(y, pred): 
    return math.sqrt(sum((y-pred)**2))/len(pred)
    
train, test = split_data(boston_rides)
train, val = split_data(train)
```

```python
train_x, train_y = labels_data(train)
val_x, val_y = labels_data(val)

vectorizer = DictVectorizer( sparse = False )
train_x = vectorizer.fit_transform(train_x.to_dict( orient = 'records' ))
#vectorizer.get_feature_names()
```

```python
train_x = np.array(train_x, dtype= float)
classifier= tree.DecisionTreeRegressor()
classifier = classifier.fit(train_x, train_y.as_matrix())
```

```python
rmse(train_y, classifier.predict(train_x))
```

```python
val_x = vectorizer.transform(val_x.to_dict(orient= 'records'))
```

```python
rmse(val_y, classifier.predict(val_x))
```

 ## 1.2 Depth Adjustment for Decision Tree Classifier

```python
train_x = np.array(train_x, dtype= float)
classifier= tree.DecisionTreeRegressor(max_depth = 8)
classifier = classifier.fit(train_x, train_y.as_matrix())
```

```python
rmse(train_y, classifier.predict(train_x))
```

```python
rmse(val_y, classifier.predict(val_x))
```

```python
# import export_graphviz 
from sklearn.tree import export_graphviz  
  
# export the decision tree to a tree.dot file 
# for visualizing the plot easily anywhere 
export_graphviz(classifier, out_file ='tree.dot', 
               feature_names =vectorizer.get_feature_names())
```
